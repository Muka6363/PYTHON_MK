Bu dosya yapısı <font color=darkred>lokal çalışma dizininde</font> de oluşturulabilir.

<br>Zipli dosya üzerine sağ tıklanarak <font color=darkred>"copy path"</font> yapılarak dosya yolu kopyalanır ve <font color=darkred>"path_to_zip_file"</font> yazan yere bu yapıştırılır.
<br>Alttaki satırda ise ziplenmiş dosyanın açılacağı <font color=darkred>directory</font> (dosya ismi hariç dosya yolu yapıştırılarak) yazılır.

```python
import zipfile
with zipfile.ZipFile("path_to_zip_file","r") as zip_ref:
    zip_ref.extractall("directory_to_extract_to")
```
